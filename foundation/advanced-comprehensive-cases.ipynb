```python
movie_genres = [i.split('|') for i in movie['genres']]
```

```python
movie_genres
```

```python
[j for i in movie_genres for j in i]
```

```python
genres = np.unique([j for i in movie_genres for j in i])
```

```python
genres.size
```

```python
counter = pd.DataFrame(np.zeros(shape=(5043,26), dtype='int32'), columns=genres)
```

```python
counter.head()
```

```python
movie_genres[0]
```

```python
counter.columns.get_indexer(movie_genres[0])
```

```python
for i in range(5043):
    counter.iloc[i, counter.columns.get_indexer(movie_genres[i])] = 1
```
