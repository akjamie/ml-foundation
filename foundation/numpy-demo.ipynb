```python
import time
import random

# generate array
py_array = []
for i in range(100000000):
    py_array.append(random.random())

nd_array = np.array(py_array)
```

```python
len(nd_array)
```

```python
# python native approach to do the sum
start_time = time.time()
a = sum(py_array);
end_time = time.time()
time_elapsed = end_time - start_time
print(f'Time elapased: {time_elapsed}')
```

```python
# numpy approach to do the sum
start_time = time.time()
a = np.sum(nd_array);
end_time = time.time()
time_elapsed = end_time - start_time
print(f'Time elapased: {time_elapsed}')
```
