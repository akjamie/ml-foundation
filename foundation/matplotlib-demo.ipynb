```python
import matplotlib.pyplot as plt
import random

x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
y_xian = [random.uniform(-8,10) for i in x]

plt.figure(figsize=(12,6),dpi=80)
plt.plot(x,y_shanghai, color='r', label='Shanghai')
plt.plot(x, y_xian, color='y', label="Xi'an")

xticks_label = ["18:{:02d}".format(i) for i in x]
plt.xticks(x[::5], xticks_label[::5])
# 5 is the step length
plt.yticks(range(0, 35, 5))

plt.grid(linestyle="--", alpha=0.3)
plt.title('Weather temporature trends')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
y_xian = [random.uniform(-8,10) for i in x]

plt.figure(figsize=(12,6),dpi=80)
#plt.plot(x,y_shanghai, color='r', label='Shanghai')
#plt.plot(x, y_xian, color='y', label="Xi'an")
figure, axes = plt.subplots(nrows=1,ncols=2,figsize=(20,8), dpi=80)

axes[0].plot(x,y_shanghai, color='r', label='Shanghai')
axes[1].plot(x, y_xian, color='y', label="Xi'an")

axes[0].legend()
axes[1].legend()

xticks_label = ["18:{:02d}".format(i) for i in x]
axes[0].set_xticks(x[::5], xticks_label[::5])
axes[0].set_yticks(range(0, 35, 5))

axes[1].set_xticks(x[::5], xticks_label[::5])
axes[1].set_yticks(range(0, 35, 5))

axes[0].grid(linestyle="--", alpha=0.3)
axes[1].grid(linestyle="-.", alpha=0.3)

axes[0].set_title('Weather temporature trends - Shanghai')
axes[1].set_title('Weather temporature trends - Xian')

axes[0].set_xlabel('Time')
axes[1].set_xlabel('Time')

axes[0].set_ylabel('Temperature')
axes[1].set_ylabel('Temperature')

plt.show()
```
