# Titanic 乘客生存预测

```python
import pandas as pd
```

```python
path = "titanic.txt"
titanic_data = pd.read_csv(path)
```

```python
titanic_data
```

```python
x = titanic_data[['pclass', 'age', 'sex']]
```

```python
y = titanic_data['survived']
```

```python
y.head()
```

```python
# 缺失值处理
#x['age'].fillna(x['age'].mean(), inplace=True)
titanic_data.fillna({'age': x['age'].mean()}, inplace=True)
```

```python
x
```

```python
x = x.to_dict(orient='records')
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=22)
```

```python
from sklearn.feature_extraction import DictVectorizer
transfer = DictVectorizer()
x_train = transfer.fit_transform(x_train)
x_test = transfer.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier,export_graphviz

estimator = DecisionTreeClassifier(criterion="entropy", max_depth=5, random_state=22)
estimator.fit(x_train, y_train)

# model training
# predict the test data
y_predict = estimator.predict(x_test)
print(f'The predicted result is: \n {y_predict}')
print(f'The actual result is: \n {y_test}')
print(f'Compare the predicted result with the actual result: \n {y_predict == y_test}')

print(f'The accuracy of the naive bayes algorithm is: \n {estimator.score(x_test, y_test)}')

# export the decision tree
export_graphviz(estimator, out_file="titanic_decison_tree.dot", feature_names=transfer.get_feature_names_out())
```
