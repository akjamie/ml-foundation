# 案例： 探究用户对物品类别的喜好细分
用户                  物品类别
user_id               aisle
1) 需要将user_id 和aisle放在同一个表中 -- 合并
2) 找到user_id 和aisle -- 交叉表和透视表
3) 特征冗余过多 -- PCA 降维

```python
import pandas as pd
```

```python
orders_products = pd.read_csv('order_products__prior.csv')
products = pd.read_csv('products.csv')
aisles = pd.read_csv('aisles.csv')
orders = pd.read_csv('orders.csv')
```

```python
orders_products.head()
```

```python
products.head()
```

```python
aisles.head()
```

```python
orders.head()
```

```python
tab1 = pd.merge(orders_products,pd.merge(aisles, products, on=['aisle_id','aisle_id']), how='left',on=['product_id'])
```

```python
tab1.head()
```

```python
tab2 = pd.merge(tab1, orders, on=['order_id','order_id'])
```

```python
tab3 = pd.crosstab(tab2['user_id'], tab2['aisle'])
```

```python
tab3.head()
```

```python
from sklearn.decomposition import PCA
```

```python
transformer = PCA(n_components=0.95)
data = transformer.fit_transform(tab3)
```
